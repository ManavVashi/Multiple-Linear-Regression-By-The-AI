# ***using AI's Linear Regression class***

# B = (X . X^T)^-1 . X^T .Y

```python
import numpy as np
```

```python
class AI_Linear_Regression:
    def __init__(self):
        self.coef_=None
        self.intercept_=None
        self.B=None
        
    def fit(self,x,y):
        x=np.insert(x,0,1,axis=1)
        self.B = np.linalg.inv(np.dot(x.T,x)).dot(x.T).dot(y)
        print(self.B)
        self.coef_=self.B[1:]
        self.intecept_=self.B[0]
        
    def predict(self,x):
        y_predict = np.dot(x,self.coef_) + self.intercept_
        print(y_predict)
    
# defining my own Linear Regression class capable to do simple as well as multiple linear regression
```
