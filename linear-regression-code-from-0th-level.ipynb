#  Y^ = X * B
1. y^ is the matrix containing predicted values of dependent variable
> size of y^ is n * 1 where n is the no of rows/persons/indivudual
> in our case it is 442 * 1
2. X is the matrix contain the values of independent variables 
> size of X matris is n * m where n is the no of rows/ persons/ individual and m is the dimention of data
> in our case n=442 and m=11 (total no of columns in our dataset including y)   so it is   442 * 11
3. B is the matrix that was predicted by model containing bius(B0) and weights(B1,B2,B3...,Bm) size is m * 1
> in our case size of B is 11 * 1

* **For calculate B we derived the formula: **
# B = (X * X^T)^-1 (X^T * Y)

***FOR DERIVATION VISIT MY GITHUB REPOSITORY WHRE YOU CAN FIND THE IDEA ABOUT IT IN "Backend Of Linear Regression" FILE.***

```python
# for finding transpose: X^T = X.T
# for finding inverse: numpy.linalg.inv() function is used
# for finding scaler product of metrics we use np.dot() fuction
```

* as we see in the "Backend Of Linear Regression" there always on extra column instead of inpupt column whose value is 1.
* it is important because constant/bius (B0) is muliply with 1 and help to get the equeation y^1 = B0 + B1x11 + B2X12 + ...
* but in our data set column containing that 1's is not present so we have to add in out AI_Linear_Regression class during model fitting

* Below is the intution of How we add 1's value containing column at index 0.

```python
import numpy as np
```

```python
print(x)

#before adding extra column
```

```python
np.insert(x,0,1,axis=1)

#axis = 0 means x axis and axis = 1 means y axis
#note:- np.insert() function not change original data, so don't worry it not modifies x
```

```python
class AI_Linear_Regression:
    
    def __init__(self):
        self.coef_ = None
        self.intercept_ = None
        self.B = None
        
    def fit(self,x,y):
        x=np.insert(x,0,1,axis=1)
        self.B = np.linalg.inv(np.dot(x.T,x)).dot(x.T).dot(y)
        self.intercept_ = self.B[0]
        self.coef_ = self.B[1:]
    
    def predict(self,x):
        y_predict=np.dot(x,self.coef_) + self.intercept_
        print(y_predict)
        
    
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=11)

#split training data and testing data using train_test_split function of odel_selection class provided by sci-kit librabry.
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model=AI_Linear_Regression()

#making object of our class
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)

#testing our model
```
