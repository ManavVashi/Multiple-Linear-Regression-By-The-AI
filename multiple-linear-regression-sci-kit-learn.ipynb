```python
#with the help of make_regression present in dataset class of sci-kit learn we can make our own dataset

import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
from sklearn.datasets import make_regression
x,y=make_regression(n_samples=100, n_features=2, n_informative=2, n_targets=1, noise=50, random_state=3)

#it generates dataset for us with 100 samples 2 input columns and 1 outpur column
```

1. n_samples: total no of rows required in the dateset
2. n_feature: total no of input column required in the dataset
3. n_informative: no of column which are atually contain useful feature for our model
4. n_target: total no of output column required in the dataset
5. random_state: same no of samples picked up every time
6. noise: standerd deviation of gaussian noise applied to the output

Let's do prediction of dependent variable using sklearn.linear_model.LinearRegression

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)

#here, value of the random state is dimenion of the dataset. In other word it is equal to no of columns in our dataset
```

```python
from sklearn.linear_model import LinearRegression

#importing LinearRegrssion class form liner_model provided by sci-kit learn
```

```python
lr=LinearRegression()
lr.fit(x,y)

#model training
```

```python
value1=lr.predict([[ 0.74505627,  1.97611078]])
value2=lr.predict([[-0.86230849,  1.07168393]])
value3=lr.predict([[-0.58850162, -1.54655582]])

print(value1)
print(value2)
print(value3)
```

```python
y_predict=lr.predict(x_test)
print(y_predict)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

#importing error function
```

```python
print("mean_squared_error= ",mean_squared_error(y_test,y_predict))
print("mean_absolute_error= ",mean_absolute_error(y_test,y_predict))
print("r2_score= ",r2_score(y_test,y_predict))
```
